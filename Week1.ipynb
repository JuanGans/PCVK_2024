```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # untuk menampilkan gambar di Google Colab
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# membuat list untuk menyimpan url dari beberapa image
urls = [
    "https://i.ibb.co.com/DPd0fTs/jiber-chan-zeeeberpunk.jpg",
    "https://asset-a.grid.id//crop/0x0:0x0/700x465/photo/2022/12/05/gunung-tertinggi-di-indonesiajp-20221205124551.jpg",
    "https://i.ibb.co.com/5TwrHgL/wallpaperflare-com-wallpaper-22.jpg"
]

# baca tampilan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((image, image_2))
    cv2_imshow(final_frame)
    print('\n')

# read image
# resize image to half size
# convert color to RGB
# concatenate image
# show image
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Daftar URL gambar
urls = [
    "https://i.ibb.co.com/jDhjdms/ktp2.jpg"
]

# Fungsi untuk menampilkan gambar dari URL
def display_image_from_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    # Tampilkan gambar
    plt.imshow(img)
    plt.axis('off')  # Menyembunyikan axis
    plt.show()

# Loop melalui URL gambar dan tampilkan gambar pertama
for url in urls:
    display_image_from_url(url)
    break  # Hanya menampilkan gambar pertama
```

```python
import requests
from PIL import Image, ImageDraw
from io import BytesIO
import matplotlib.pyplot as plt

# Daftar URL gambar
urls = [
    "https://i.ibb.co.com/jDhjdms/ktp2.jpg"
]

# Fungsi untuk menampilkan gambar dari URL dengan area yang disensor
def display_and_blur_image_from_url(url):
    # Ambil gambar dari URL
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    # Koordinat dan ukuran untuk area yang akan disensor
    positions = [
        (175, 135),  # Posisi untuk area NIK (x1, y1)
        (175, 155),  # Posisi untuk area Alamat (x1, y1)
    ]

    sizes = [
        (200, 15),  # Ukuran untuk area NIK (width, height)
        (80, 15),  # Ukuran untuk area Alamat (width, height)
    ]

    # Warna balok untuk menyensor
    block_colors = [
        (255, 255, 255),  # Putih untuk area pertama
        (0, 0, 0),        # Hitam untuk area kedua
    ]

    # Buat objek ImageDraw untuk menggambar balok pada gambar
    draw = ImageDraw.Draw(img)

    # Gambar balok untuk menyensor dengan warna yang sesuai
    for i in range(len(positions)):
        x1, y1 = positions[i]
        width, height = sizes[i]
        x2 = x1 + width
        y2 = y1 + height
        draw.rectangle([x1, y1, x2, y2], fill=block_colors[i])

    # Tampilkan gambar yang sudah disensor
    plt.imshow(img)
    plt.axis('off')  # Menyembunyikan axis
    plt.show()

# Loop melalui URL gambar dan tampilkan gambar pertama
for url in urls:
    display_and_blur_image_from_url(url)
    break  # Hanya menampilkan gambar pertama
```
