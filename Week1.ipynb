```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # untuk menampilkan gambar di Google Colab
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# membuat list untuk menyimpan url dari beberapa image
urls = [
    "https://i.ibb.co.com/DPd0fTs/jiber-chan-zeeeberpunk.jpg",
    "https://asset-a.grid.id//crop/0x0:0x0/700x465/photo/2022/12/05/gunung-tertinggi-di-indonesiajp-20221205124551.jpg",
    "https://i.ibb.co.com/5TwrHgL/wallpaperflare-com-wallpaper-22.jpg"
]

# baca tampilan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((image, image_2))
    cv2_imshow(final_frame)
    print('\n')

# read image
# resize image to half size
# convert color to RGB
# concatenate image
# show image
```

```python

# Mendapatkan dimensi gambar
tinggi = image_2.shape[0]
lebar = image_2.shape[1]

# Mencetak resolusi gambar
print("Resolusi image: tinggi x lebar = ", tinggi, "x", lebar)

# Menampilkan gambar
cv2_imshow(image_2)
```

```python
# Konversi gambar ke format RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal di tengah gambar
for x in range(tinggi):
    image_3[x, int(lebar / 2)] = [255, 255, 255]

# Membuat garis diagonal menyilang
for y in range(lebar):
    image_5[int(tinggi / lebar * y), y] = [255, 255, 255]
    image_5[int(tinggi / lebar * y), -y] = [255, 255, 255]

# Menggabungkan gambar secara horizontal
final_frame = cv.hconcat((image_2, image_3))
final_frame2 = cv.hconcat((image_4, image_5))

# Menampilkan gambar
cv2_imshow(final_frame)
cv2_imshow(final_frame2)
```

```python
# Konversi gambar ke format RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal berwarna putih di bagian tertentu
for y in range(lebar):
    if y > 100 and y < 200:
        image_3[int(tinggi / 2), y] = [255, 255, 255]

# Menggabungkan gambar secara horizontal
final_frame = cv.hconcat((image_2, image_3))

# Menampilkan gambar
cv2_imshow(final_frame)
```

```python
# Konversi gambar ke format RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Mendapatkan dimensi gambar
tinggi, lebar, _ = image_3.shape

# Membuat kotak dengan pixel berwarna putih, dengan pengecekan batas
x_start, x_end = 100, 200
y_start, y_end = 100, 200

x_end = min(x_end, tinggi)  # Memastikan x_end tidak melebihi tinggi gambar
y_end = min(y_end, lebar)   # Memastikan y_end tidak melebihi lebar gambar

for y in range(y_start, y_end):
    for x in range(x_start, x_end):
        image_3[x, y] = [255, 255, 255]

# Menggabungkan gambar secara horizontal
final_frame = cv.hconcat((image_2, image_3))

# Menampilkan gambar
cv2_imshow(final_frame)
```

```python
from PIL import Image, ImageDraw

# Buka gambar
img = Image.open('ktp.png')

# Buat objek ImageDraw
draw = ImageDraw.Draw(img)

# Koordinat untuk area yang akan disensor
# Gantilah koordinat dengan nilai yang sesuai untuk gambar kamu
areas_to_blur = [
    (50, 100, 200, 170),  # Area NIK
    (50, 200, 200, 220),  # Area Alamat
    (50, 250, 200, 270)   # Area Berlaku Hingga
]

# Warna balok untuk menyensor (misalnya putih)
block_color = (255, 255, 255)

# Gambar balok untuk menyensor
for area in areas_to_blur:
    draw.rectangle(area, fill=block_color)

# Simpan gambar yang sudah disensor
img.save('ktp_sensored.png')

img.show() # tampilkan gambar
display(img) # metode alternatif tampilkan gambar
```
