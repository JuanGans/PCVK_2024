```python
black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(black_img)
```

```python
# Perhatikan koordinat titik-titik pt1 dan pt2
cv.rectangle(black_img, pt1=(384, 8), pt2=(510, 150), color=(8, 255, 0), thickness=10)
plt.imshow(black_img)
```

```python
# Menggambar sebuah kotak dengan koordinat pt1 dan pt2
cv.rectangle(black_img, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=15)
plt.imshow(black_img)
```

```python
# Menggambar sebuah lingkaran dengan pusat dan radius yang ditentukan
cv.circle(black_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)
```

```python
# Menggambar sebuah garis dari pt1 ke pt2
cv.line(black_img, pt1=(0, 0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)
```

```python
# Mengatur font
font = cv.FONT_HERSHEY_SIMPLEX

# Menambahkan teks pada gambar
cv.putText(black_img, text="Hello", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 6), thickness=2, lineType=cv.LINE_AA)

# Menampilkan gambar dengan teks
plt.imshow(black_img)
```
